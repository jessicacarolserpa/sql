```sql
SELECT ProductID, name, StandardCost, listprice
FROM product p
WHERE not EXISTS
(SELECT *
FROM SalesOrderDetail sod
WHERE p.productid = sod.productid)
```

Exerc√≠cio 6 Liste o c√≥digo das categorias, o c√≥digo dos produtos e o somat√≥rio de vendas c√≥digo de categoria, c√≥digo de produto

```sql
SELECT p.ProductCategoryID, p.productid, sum(sod.linetotal) as total 
FROM product p 
JOIN SalesOrderDetail sod
on p.productid = sod.ProductID
GROUP BY p.ProductCategoryID, p.productid
order by total desc
```
